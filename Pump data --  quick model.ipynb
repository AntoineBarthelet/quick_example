```python
pumps['dow'] = pumps['datetime'].apply(lambda x: x.dayofweek)
pumps['day'] = pumps['datetime'].apply(lambda x: x.day)
pumps['moy'] = pumps['datetime'].apply(lambda x: x.month)
```

```python
import sklearn
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
sklearn.__version__
```

```python
pumps['HP_1 - 1month'] = pumps['HP_1'].shift(int(-1440/5)*30)
pumps['HP_2 - 1month'] = pumps['HP_2'].shift(int(-1440/5)*30)
pumps['HP_3 - 1month'] = pumps['HP_3'].shift(int(-1440/5)*30)
```

```python
pumps['HP_1 - 1month'] = pumps['HP_1 - 1month'].apply(lambda x: 1 if x>10 else 0)
pumps['HP_2 - 1month'] = pumps['HP_2 - 1month'].apply(lambda x: 1 if x>10 else 0)
pumps['HP_3 - 1month'] = pumps['HP_3 - 1month'].apply(lambda x: 1 if x>10 else 0)
```

```python
pumps['Head us - 1month'] = pumps['Head us'].shift(int(-1440/5)*30)
pumps['Head us - 3month'] = pumps['Head us'].shift(int(-1440/5)*90)
```

```python
pumps.head()
```

```python
pumps['MGD Flow - 1month'] = pumps['MGD Flow'].shift(int(-1440/5)*30)
```

```python
# wet welll levels (ultrasonice)

# regr = MLPRegressor(hidden_layer_sizes=(10,10))
regr = RandomForestRegressor(n_estimators = 50)
train = pumps[pumps['datetime']<pd.to_datetime('2016-06-01')].dropna()
test = pumps[pumps['datetime']>=pd.to_datetime('2016-06-01')].dropna()
features = ['dow', 'day', 'moy', 'Head us',
            'HP_1', 'HP_2', 'HP_3',
            'MGD Flow']
label = 'Head us - 3month'
```

```python
regr.fit(train[features], train[label])
```

```python
preds = regr.predict(test[features])
test['preds'] = preds
test['mae'] = np.abs(test['preds'] - test[label])
```

```python
fig = plt.figure(figsize=[12,8])
plt.plot(test['datetime'], test[label], label='actual head')
plt.plot(test['datetime'], preds, label='predicted head')
plt.title('Predictions vs Actual Head - 3 month out, MAE: ' + str(test['mae'].mean())[:4])
plt.legend()
```
